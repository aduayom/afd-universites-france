```python
# Utilisation d'un contexte de style Seaborn temporaire pour cette cellule uniquement
with sns.axes_style("whitegrid"):
    # Cr√©ation d'une figure avec 2 sous-graphiques (1 ligne, 2 colonnes)
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))  # 1 ligne, 2 colonnes

    # Liste des colonnes √† visualiser
    colonnes = ['order_channel', 'last_order_channel']

    # Parcours des colonnes et cr√©ation des histogrammes tri√©s
    for i, col in enumerate(colonnes):
        ax = axes[i]  # Utiliser l'axe correspondant
        
        # Calculer les valeurs de fr√©quence tri√©es pour chaque colonne
        sorted_order = Data_CRM[col].value_counts().index
        
        total = float(len(Data_CRM[col]))
        
        # Tracer les courbes avec les cat√©gories tri√©es
        sns.countplot(x=col, data=Data_CRM, palette="viridis", order=sorted_order, ax=ax)
        ax.set_title(f'Distribution de {col}')
        
        # Ajout des pourcentages sur les barres
        for p in ax.patches:
            height = p.get_height()
            ax.text(p.get_x() + p.get_width()/2.,
                    height + 3,
                    '{:1.2f}%'.format(100 * height / total),
                    ha="center")

    # Ajustement de l'espace entre les graphiques
    plt.tight_layout()
    plt.show()
```

    <h3 style="color: #75DFC1;">üìä Introduction</h3>
    <p>
        Afin d'enrichir notre analyse, nous avons cr√©√© de nouvelles variables qui apportent des informations suppl√©mentaires sur les comportements d'achat de nos clients. Ces variables sont d√©riv√©es des colonnes de dates, telles que les dates des premi√®res et derni√®res commandes, ainsi que les commandes effectu√©es en ligne et hors ligne. Cela nous permet de mieux comprendre l'engagement des clients et leurs pr√©f√©rences en termes de canaux.
    </p>
    <h4>üïí Dur√©e de la Relation Client (Customer Lifetime)</h4>
    <p>
        <strong>Variable :</strong> <code>customer_lifetime</code> <br>
        <strong>Description :</strong> Cette variable mesure la dur√©e totale (en jours) entre la premi√®re commande du client et sa derni√®re commande, indiquant depuis combien de temps le client est actif avec l'entreprise. Elle permet d'√©valuer la fid√©lit√© des clients au fil du temps.
    </p>
    <h4>üîÑ Temps Moyen entre les Commandes (Time Between Orders)</h4>
    <p>
        <strong>Variable :</strong> <code>time_between_orders</code> <br>
        <strong>Description :</strong> Cette variable repr√©sente le temps moyen (en jours) entre chaque commande effectu√©e par le client. Elle est calcul√©e en divisant la dur√©e de la relation client par le nombre total de commandes. Elle aide √† comprendre la fr√©quence d'achat du client.
    </p>
    <h4>üìÖ R√©cence en Ligne vs Hors Ligne (Online vs Offline Recency)</h4>
    <p>
        <strong>Variable :</strong> <code>online_vs_offline_recency</code> <br>
        <strong>Description :</strong> Cette variable compare la r√©cence des commandes en ligne par rapport aux commandes hors ligne en calculant la diff√©rence en jours entre la derni√®re commande en ligne et la derni√®re commande hors ligne. Elle permet d'analyser quel canal est privil√©gi√© par le client dans ses achats r√©cents.
    </p>
    <h4>‚öñÔ∏è Fr√©quence des Commandes en Ligne vs Hors Ligne (Online vs Offline Order Frequency)</h4>
    <p>
        <strong>Variable :</strong> <code>online_vs_offline_order_frequency</code> <br>
        <strong>Description :</strong> Cette variable exprime le ratio entre le nombre total de commandes en ligne et hors ligne. Elle permet de d√©terminer si un client pr√©f√®re commander en ligne ou hors ligne et d'√©valuer ses habitudes d'achat par canal.
    </p>
    <p>
        Ces nouvelles variables sont cruciales pour mieux comprendre les habitudes d'achat de nos clients et pour adapter les strat√©gies de marketing en fonction de leurs pr√©f√©rences. Elles nous permettront d'identifier des segments sp√©cifiques de clients, tels que ceux qui sont plus fid√®les √† un canal d'achat, et de personnaliser les campagnes marketing en cons√©quence.
    </p>
</div>


```python
# 1. Cr√©ation de la variable 'customer_lifetime'
# Dur√©e de la relation client : nombre de jours entre la premi√®re commande et la derni√®re commande
Data_CRM['customer_lifetime'] = (Data_CRM['last_order_date'] - Data_CRM['first_order_date']).dt.days

# 2. Cr√©ation de la variable 'time_between_orders'
# Temps moyen entre les commandes
# On utilise le nombre total de commandes (en ligne + hors ligne) pour obtenir la fr√©quence moyenne d'achat
Data_CRM['total_orders'] = Data_CRM['order_num_total_ever_online'] + Data_CRM['order_num_total_ever_offline']
# √âviter la division par 0 en v√©rifiant que total_orders n'est pas nul
Data_CRM['time_between_orders'] = Data_CRM['customer_lifetime'] / Data_CRM['total_orders']
Data_CRM['time_between_orders'] = Data_CRM['time_between_orders'].replace([float('inf'), -float('inf')], None)

# 3. Cr√©ation de la variable 'online_vs_offline_recency'
# Diff√©rence en jours entre la derni√®re commande en ligne et la derni√®re commande hors ligne
Data_CRM['online_vs_offline_recency'] = (Data_CRM['last_order_date_online'] - Data_CRM['last_order_date_offline']).dt.days

# 4. Cr√©ation de la variable 'online_vs_offline_order_frequency'
# Ratio entre le nombre total de commandes en ligne et hors ligne
Data_CRM['online_vs_offline_order_frequency'] = Data_CRM['order_num_total_ever_online'] / Data_CRM['order_num_total_ever_offline']

Data_CRM[['master_id','customer_lifetime','time_between_orders','online_vs_offline_recency','online_vs_offline_order_frequency']].head()
```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np
import pandas as pd
from random import sample, uniform

def hopkins(X):
    d = X.shape[1]  # Nombre de dimensions
    n = len(X)  # Nombre d'√©chantillons
    m = int(0.3 * n)  # Proportion d'√©chantillons pour le test
    
    # Initialiser NearestNeighbors
    nbrs = NearestNeighbors(n_neighbors=2).fit(X)
    
    # Variables pour stocker les distances
    ujd = []
    wjd = []
    
    # G√©n√©rer des points al√©atoires dans l'espace des donn√©es
    for j in range(m):
        # G√©n√©rer un point al√©atoire dans les limites des colonnes (dimensions)
        random_point = [uniform(np.min(X[:,i]), np.max(X[:,i])) for i in range(d)]
        
        # Calculer les distances des points al√©atoires
        u_dist, _ = nbrs.kneighbors([random_point], n_neighbors=2, return_distance=True)
        ujd.append(u_dist[0][1])
        
        # Choisir un point r√©el al√©atoire dans X et calculer la distance
        w_dist, _ = nbrs.kneighbors(X[np.random.randint(0, n)].reshape(1, -1), n_neighbors=2, return_distance=True)
        wjd.append(w_dist[0][1])
    
    # Calcul de la statistique de Hopkins
    H = sum(ujd) / (sum(ujd) + sum(wjd))
    
    return H
```

<h3 style="color: #75DFC1;">üìä Impl√©mentation du Mod√®le de Clustering</h3> 

<p> Dans cette section, nous allons impl√©menter plusieurs approches de clustering afin d'explorer les diff√©rentes mani√®res de segmenter nos donn√©es. Chaque m√©thode a ses sp√©cificit√©s et son int√©r√™t, et en les comparant, nous pourrons identifier celle qui offre la meilleure segmentation. Nous allons tester les m√©thodes suivantes : </p> 
<p> <strong>K-Means :</strong> Algorithme de clustering tr√®s populaire qui partitionne les donn√©es en fonction de la distance euclidienne aux centres de clusters. Son principal int√©r√™t r√©side dans sa simplicit√© et son efficacit√©, mais il est limit√© par sa sensibilit√© aux outliers et ne fonctionne bien que pour des clusters de forme sph√©rique. </p> 
<p> <strong>K-Prototypes :</strong> Extension du K-Means permettant de g√©rer √† la fois des variables num√©riques et cat√©gorielles. Il est particuli√®rement utile lorsque l'on travaille avec des donn√©es mixtes, ce qui est souvent le cas dans les analyses CRM. </p> <p> <strong>K-Mix :</strong> Une autre approche adapt√©e aux donn√©es contenant des variables continues et discr√®tes. K-Mix utilise des m√©langes de gaussiennes et est capable de mod√©liser des donn√©es plus complexes avec des distributions plus vari√©es. </p>
<p> <strong>Agglomerative Clustering :</strong> Technique de clustering hi√©rarchique qui construit une hi√©rarchie de clusters en fusionnant successivement les plus proches. Ce type de clustering est int√©ressant car il permet de visualiser les relations hi√©rarchiques entre les clusters, mais il est souvent plus lent que K-Means pour de grandes quantit√©s de donn√©es. </p> 
<p> L'objectif de cette √©tape est de tester et de comparer ces diff√©rentes approches afin de trouver celle qui segmente le mieux nos donn√©es clients en fonction des caract√©ristiques disponibles. Cette analyse nous permettra d'exploiter au mieux les r√©sultats du clustering pour ajuster nos strat√©gies marketing. </p> 

</div>

<div style="font-family: 'Arial', sans-serif; color: #333; font-size: 14px;"> 
<h3 style="color: #75DFC1;">üî¢  M√©thodologie pour S√©lectionner le Nombre de Clusters</h3> 
<p> La d√©termination du nombre optimal de clusters est une √©tape cl√© dans la mise en place de nos mod√®les de clustering. Pour identifier le nombre de clusters le plus adapt√© √† nos donn√©es, nous nous appuyons sur deux m√©thodes compl√©mentaires : la m√©thode du coude et le score de silhouette. Ces approches permettent de trouver un √©quilibre entre la simplicit√© de la segmentation et la qualit√© des regroupements. </p>

<p> <strong>M√©thode du Coude :</strong> Cette m√©thode consiste √† √©valuer l'√©volution de l'<strong>inertie</strong> (somme des distances au carr√© des points par rapport √† leur centre de cluster) en fonction du nombre de clusters. L'inertie diminue lorsque le nombre de clusters augmente, mais √† partir d'un certain point, le gain de r√©duction devient marginal. Le "coude" de la courbe repr√©sente ce point, o√π une augmentation suppl√©mentaire du nombre de clusters n'apporte plus de r√©elle am√©lioration de la segmentation. </p> 

<p> <strong>Score de Silhouette :</strong> Ce score √©value la qualit√© de la segmentation en mesurant √† quel point les points sont proches de leur propre cluster par rapport aux autres clusters. Il prend des valeurs entre -1 et 1, o√π un score proche de 1 indique des clusters bien distincts et compacts. En comparant les scores de silhouette pour diff√©rents nombres de clusters, nous pouvons identifier celui qui offre le meilleur compromis entre s√©paration des clusters et compacit√©. </p> 

<p> Ces deux m√©thodologies sont appliqu√©es conjointement pour valider le choix du nombre de clusters. La m√©thode du coude nous aide √† identifier le point √† partir duquel l'ajout de clusters devient moins pertinent, tandis que le score de silhouette nous permet de v√©rifier la coh√©rence et la qualit√© de la segmentation choisie. </p> 

</div>

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score


# Liste pour stocker les r√©sultats
results = []

# It√©ration sur le nombre de clusters de 1 √† 10
for n_clusters in range(1, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)  # Fixer la seed pour la reproductibilit√©
    kmeans.fit(Data_normalized)
    
    # Calculer l'inertie (distorsion) : somme des distances au carr√©
    inertia = kmeans.inertia_
    
    # Calculer le score de silhouette uniquement si n_clusters > 1
    if n_clusters > 1:
        silhouette_avg = silhouette_score(Data_normalized, kmeans.labels_)
    else:
        silhouette_avg = np.nan  # Pas de score de silhouette pour un seul cluster
    
    # Ajouter les r√©sultats √† la liste
    results.append({
        'n_clusters': n_clusters,
        'inertia': inertia,
        'silhouette_score': silhouette_avg
    })

# Convertir les r√©sultats en DataFrame
df_results = pd.DataFrame(results)

# Afficher le DataFrame des r√©sultats
print(df_results)
```

```python
# Visualiser l'inertie
plt.figure(figsize=(10, 5))
plt.plot(df_results['n_clusters'], df_results['inertia'], marker='o', label='Inertia')
plt.title('Inertia en fonction du nombre de clusters')
plt.xlabel('Nombre de clusters')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()

# Visualiser le score de silhouette
plt.figure(figsize=(10, 5))
plt.plot(df_results['n_clusters'], df_results['silhouette_score'], marker='o', label='Silhouette Score', color='green')
plt.title('Score de silhouette en fonction du nombre de clusters')
plt.xlabel('Nombre de clusters')
plt.ylabel('Score de silhouette')
plt.grid(True)
plt.show()
```

L'algorithme de K-means regroupe les individus en minimisant la distance intra-cluster, ce qui permet de former des groupes compacts et distincts. </p> <p> Nous appliquons le mod√®le sur notre base de donn√©es standardis√©e, puis nous attribuons chaque individu √† un cluster sp√©cifique. Cela nous permettra de mieux comprendre les caract√©ristiques de chaque groupe et de d√©velopper des strat√©gies adapt√©es pour chaque segment de clients. </p> </div>

```python
# Choisissez le nombre optimal de clusters trouv√© pr√©c√©demment
n_clusters_optimal = 3  # Remplacez par le nombre de clusters choisi

# Instanciation et ajustement du mod√®le K-means
kmeans = KMeans(n_clusters=n_clusters_optimal, random_state=42)  # Fixer la seed pour la reproductibilit√©
kmeans.fit(Data_normalized)

# Attribution des clusters √† chaque individu
# 'labels_' contient les labels de chaque individu apr√®s le clustering
Data_CRM['cluster_Kmeans'] = kmeans.labels_

# Afficher les premi√®res lignes avec l'attribution des clusters
print(Data_CRM[['cluster_Kmeans']].head())
```

<div style="font-family: 'Arial', sans-serif; color: #333; font-size: 14px;"> <h3 style="color: #75DFC1;">‚öôÔ∏è Mise en Place du Mod√®le K-Prototypes</h3> 

<p> Le mod√®le <strong>K-Prototypes</strong> est une extension de l'algorithme K-means, con√ßu pour traiter √† la fois des variables num√©riques et des variables cat√©gorielles. Il permet de segmenter les donn√©es mixtes en combinant la distance euclidienne pour les variables continues et la distance de correspondance (matching) pour les variables qualitatives. </p>
<p> Dans cette section, nous sp√©cifions les variables continues et les variables cat√©gorielles, ainsi que le nombre optimal de clusters d√©termin√© pr√©c√©demment. Le mod√®le K-Prototypes est ensuite appliqu√© pour regrouper les individus, en tenant compte de la nature des donn√©es. </p> 

<p> L'algorithme K-Prototypes est particuli√®rement utile pour les analyses CRM o√π les informations sur les clients comprennent des caract√©ristiques √† la fois quantitatives (par exemple, le montant total des achats) et qualitatives (par exemple, le canal de commande pr√©f√©r√©). </p> 

</div>

```python
from kmodes.kprototypes import KPrototypes

# Supposons que Data_CRM soit votre DataFrame d'origine contenant les donn√©es.
# Sp√©cifiez les variables continues et les variables cat√©gorielles.
variables_continues = ['customer_lifetime', 'time_between_orders', 'online_vs_offline_recency', 'online_vs_offline_order_frequency','customer_lifetime', 'time_between_orders', 'online_vs_offline_recency', 'online_vs_offline_order_frequency','order_num_total_ever_online', 'order_num_total_ever_offline', 
             'customer_value_total_ever_offline', 'customer_value_total_ever_online']
variables_categoriques = ['order_channel', 'last_order_channel']

# S√©lection des donn√©es pour le clustering
data_kproto = Data_CRM[variables_continues + variables_categoriques]

# Remplacement des valeurs manquantes √©ventuelles (facultatif)
# data_kproto = data_kproto.fillna(0)

# Conversion des colonnes cat√©gorielles en type 'object' pour K-Prototypes
for col in variables_categoriques:
    data_kproto[col] = data_kproto[col].astype(str)

# D√©finir le nombre de clusters
n_clusters_optimal = 3  # Remplacez par le nombre de clusters souhait√©

# Instanciation et ajustement du mod√®le K-Prototypes
kproto = KPrototypes(n_clusters=n_clusters_optimal, random_state=42)
clusters = kproto.fit_predict(data_kproto, categorical=[data_kproto.columns.get_loc(col) for col in variables_categoriques])

# Ajouter les attributions de clusters √† la DataFrame d'origine
Data_CRM['cluster_KPrototypes'] = clusters

# Afficher les premi√®res lignes avec les attributions de clusters
print(Data_CRM[['cluster_KPrototypes'] + variables_continues + variables_categoriques].head())
```
